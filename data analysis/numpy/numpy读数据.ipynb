```python
import numpy as np
from matplotlib import pyplot as plt


uk_file_path = "../database/GB_video_data_numbers.csv"

# 导入数据
t_uk = np.loadtxt(uk_file_path,delimiter=",",dtype="int")

# unpack=true则为转置导入
# t1 = np.loadtxt(us_file_path,delimiter=",",dtype="int",unpack=True)

# 取出来英国的youtube评论数和喜欢数,第1列是like数，第3列是comment数
# 为了方便作图观察，筛选like数小于等于200000的视频
t_uk=t_uk[t_uk[:,1]<=250000]
print(t_uk.shape)

# 取第一列 like
t_uk_like=t_uk[:,1]
# 取最后一列 comment
t_uk_comment = t_uk[:,-1]

plt.figure(figsize=(20,8),dpi=80)
#打印横轴纵轴标签
plt.xlabel("like")
plt.ylabel("comment")

#画散点图
plt.scatter(t_uk_like,t_uk_comment)

plt.show()
```
