## 折线图plot

### x轴、y轴和刻度

```python
x=range(2,26,2)
y=[random.randint(1,10)for i in x]
plt.figure(figsize=(20,8),dpi=80)
#x轴刻度
x_ticks_label=[f"{i}:00"for i in x]
#x为刻度数量、x_ticks_label为刻度标签样式、rotation旋转角度
plt.xticks(x,x_ticks_label,rotation=45)
y_ticks_label=[f"{i}°C"for i in range(min(y),max(y)+1)]
plt.yticks(range(min(y),max(y)+1),y_ticks_label)
plt.plot(x,y)
plt.show()
```

### 多线和图例

### 划分子图

```python
x = np.arange(1, 100)
fig=plt.figure(figsize=(20,10),dpi=80)
fig,axes=plt.subplots(2,2)
#子图位置编号
ax1=axes[0,0]
ax2=axes[0,1]
ax3=axes[1,0]
ax4=axes[1,1]
fig=plt.figure(figsize=(20,10),dpi=80)
#图 1
ax1.plot(x, x)
#图 2
ax2.plot(x,-x)
#图 3
ax3.plot(x, x**2)
ax3.grid(color='r', linestyle='--', linewidth=1,alpha=0.3)
#图 4
ax4.plot(x, np.log(x))
plt.show()
```

## 条形图（柱状图）bar

```python
from matplotlib import pyplot as plt
from matplotlib import font_manager
my_font = font_manager.FontProperties(fname="C:/Windows/Fonts/simhei.ttf",size=16)
a = ['北京','上海','广州','深圳','天津','河北']
b = [38.13,19.85,14.89,11.36,6.47,5.93]
plt.figure(figsize=(20,8),dpi=80)
# 绘制条形图的方法
#条形的宽度
width=0.3
#绘制条形图
rects = plt.bar(range(len(a)),b,width=0.3,color='r')
'''
绘制条形图后，会返回一个包含每条柱子信息的对象，为可迭代对象
'''
#设置横坐标标签
plt.xticks(range(len(a)),a,fontproperties=my_font,rotation=45)
# 在条形图上加标注(水平居中)
for rect in rects:
    #取每条柱子的高度
    height = rect.get_height()
    #rect.get_x() + rect.get_width() / 2   表示标注的位置为柱子中心点的横坐标
    #height+0.3  表示标注的位置为柱子高度之上0.3
    #ha="center"  表示水平居中(标注与给定位置的对齐位置，即数字的中心与给定位置对齐)
    #va="bottom"  表示底部对其(标注与给定位置的对齐位置，即数字的底部与给定位置对齐)
    plt.text(rect.get_x() + rect.get_width() / 2, height+0.3, '%.2f'%height,ha="center",va="bottom")
plt.show()
```
