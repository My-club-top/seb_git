```python
'''
1、 机器学习：从数据中自动分析获得规律，利用规律对位置数据进行预测

2、 数据类型：
        离散型：如数组，不可分割，用于分类
        连续性：区间，可分割，用于回归

3、 机器学习数据特征间符合独立同分布
    独立性 指每个观测值的出现不影响其他观测值，数学上表现为联合概率等于各自概率的乘积： P(X₁, X₂, …, Xₙ) = P(X₁) × P(X₂) × … × P(Xₙ)。
    同分布性 则要求所有观测值服从同一分布函数 F(x)，如正态分布、均匀分布等。

4、 特征工程是将原始数据转换为更好地代表预测模型的潜在问题的特征的过程，从而提高了模型对未知数据预测的准确性

5、 TF-IDF的主要思想是：如果某个词或短语在一篇文章中出现的概率高，并且在其他文章中很少出现，则认为此词或者短语具有很好的类别区分能力，适合用 来分类。
    tf*idf  来代表重要性程度
    Tf:term frequency:词的频率出现的次数
    idf:逆文档频率inverse document frequency  =  log(总文档数量/该词出现的文档数量)  以10为低（权重）
    在文档频率上加一来平滑 idf 权重  log(n/n(x)) -> log(n+1/n(x)+1)+1
    防止零分裂，若训练集中有某个词，测试集中没有，就会造成n(x)分母为零，log(n/n(x))出现零分裂

6、 归一化 X'=(x-min)/(max-min) 将数据映射到0到1之间
    若要其他范围，则X''=X‘*(max-min)+min
    最大值与最小值非常容易受异常点影响，所以这种方法鲁棒性较差（数据多了表现不好），只适合传统精确小数据（样本量小）场景

7、 标准化 X'=(x-均值)/标准差   （标准差 https://zhuanlan.zhihu.com/p/17830577910 ）
    方差：（数值-均值）求和再除以数据数
    可以把数据变换到均值为0,标准差为1的均匀分布，不是正态分布
    因为减去均值，所以受异常值影响小
    对于归一化来说：如果出现异常点，影响了最大值和最小值，那么结果显然会发生改变
    对于标准化来说：如果出现异常点，由于具有一定数据量，少量的异常点对于平均值的影响并不大，从而方差改变较小。

8、 fit_transform和transform的不同，fit_transform在训练完训练集后，均值标准差等确定，不会随着测试集变化
    transform和fit_transform不同是，transform用于测试集，而且不会重新找最小值和最大值,不会重新计算均值方差

9、 冗余：部分特征的相关度高，容易消耗计算性能
    噪声：部分特征对预测结果有负影响

10、特征选择方法：
    Filter(过滤式):VarianceThreshold （删除方差小的特征）
	Embedded(嵌入式)：正则化、决策树
	Wrapper(包裹式)：集成学习

11、PCA：
    目的：使数据维数压缩，尽可能降低原数据的维数（复杂度），损失少量信息。
    作用：可以削减回归分析或者聚类分析中特征的数量
    特征数量达到上百的时候，开始使用PCA

    PCA (Principal ComponentAnalysis，主成分分析)的结果可能会因为以下几个原因而不稳定:
    ①数据归一化的方式不同:对于PCA来说，数据的归一化可以采用标准化或者最大最小值归一化等方法。不同的归—化方式会导致不同的结果。特征值计算的精度:在特征值计算算精度也可能会影响结果的稳定性。
    ②特征值相同时的特征向量计算:如征向量的计算就不唯一了,不同的计算方式也会导致结果不同。
    ③随机选择初始向量:在迭代计算特征向量的过程中，需要随机选择一个初始向量，这也可能会导致结果不稳定。为了提高结果的稳定性，可以采用多次运行PCA并取平均值的方法。另外，在进行PCA之前，可以对数据进行多次重复采样，然后对每次采样得到的数据进行PCA分析，以进一步提高结果的可靠性。

12、机器学习：算法是核心，数据和算力是基础

13、 监督学习
        分类    k-近邻算法、贝叶斯分类、决策树与随机森林、逻辑回归、神经网络
        回归    线性回归、岭回归、神经网络
        标注    隐马尔可夫模型     (不做要求)
    无监督学习（没有标签值，给参数不给结果）
        聚类    k-means（有几十种聚类）

'''
```
